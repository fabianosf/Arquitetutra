## SeleÃ§Ã£o por CondicÃ£o

```python
dados.head()
```

```python
# Busca pra mim as pessoas da 1Âª classe
dados.loc[dados['Pclass'] == 1]
```

```python
# Busca pra mim as pessoas do sexo feminino
dados.loc[dados['Sex'] == 'female']
```

```python
# Busca as pessoas com idade menor que 18 anos
dados.loc[dados['Age'] < 18 ]
```

```python
# Busca pessoas que tenham a idade <= 18 e do sexo masculino
dados.loc[(dados['Age'] <=18) & (dados['Sex'] == 'male')]
```

```python
# Busca pessoas que tenham a idade <= 18 e do sexo masculino
# Alem disso me retorne tb o nome e o portao de embarque
# Filtro dentro de um filtro, Ã© um filtro mais avancado
dados.loc[(dados['Age'] <=18) & (dados['Sex'] == 'male')][['Name','Embarked']]
```

```python
# Busca pessoas que tenham a idade <= 18 e do sexo masculino
# Alem disso me retorne tb o nome e o portao de embarque
# Filtro dentro de um filtro, Ã© um filtro mais avancado
# Porem agora coloquei do sexo masculino ou feminino pq nao existe masculino e feminino...
dados.loc[(dados['Age'] <=18) & (dados['Sex'] == 'male') | (dados['Sex'] == 'female')][['Name','Embarked']]
```

## Tratamento de Dados Faltantes

```python
dados1 = dados.copy()
dados2 = dados.copy()
dados3 = dados.copy()
```

```python
# False Ã© onde nao tem dados faltantas, True Ã© onde tem dados faltantes
dados.isnull().sum()
```

```python
# 891 linha, 12 colunas
# dados.shape[0] retorna 891, dados.shape[1] retorna 12
dados.shape
```

```python
# Aqui ele vai me retornar a porcentagem
dados.isnull().sum() / dados.shape[0] * 100
```

```python
# Quais sao os dados que estao faltando dentro da coluna Age ?
dados.loc[dados['Age'].isnull()]
```

```python
# Quais sao os dados que estao faltando dentro da coluna Cabin ?
dados.loc[dados['Cabin'].isnull()]
```

## Tratando os dados agora !!

```python
# onde tive NaN ele vai apagar tudo. ESTE Ã‰ UM METODO MUITO RADICAL, NAO MTO RECOMENDADO
# USE ESSE METODO QDO TIVER POUCOS DADOS..TIPO 2, 3 ,4 DADOS FALTANDO...Ã‰ MAIS INDICADO
# ACIMA DE 5% NAO USAR O DROPNA
dados1.dropna(inplace=True)
```

```python
dados1.shape
```

```python
# Verificando agora
dados1.isnull().sum()
```

## Tratamento de dados faltantes !!! Metodos mais indicados!!!

```python
dados2.isnull().sum()
```

```python
# Vou tirar a media das idades
media_idades = dados2['Age'].mean()
print(media_idades)
```

```python
# fillna preenche os dados faltantes com algum valor
# nesse caso usarei o valor media_idades, vou arredondar pra 30
dados2['Age'].fillna(30, inplace=True)
```

```python
dados2.isnull().sum()
```

```python
# ele pega o ultimo dado valido antes do dado faltante e vai jogar pra frente
dados3['Embarked'].ffill(inplace=True)
dados3.isnull().sum()
```

```python
# Excluindo coluna
# axis = 1 Ã© coluna, axis = 0 Ã© igual a linha
dados3.drop(['Cabin'], axis=1, inplace = True)
dados3.isnull().sum()
```

```python
dados3.fillna(media_idades, inplace=True)
dados3.isnull().sum()
```
